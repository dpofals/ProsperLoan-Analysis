### The structure of the dataset

> Unwrangled,the ProsperLoan dataset has 113,937 rows and 81 columns. The information in every row corresponds to a loan that was issued by Prosper MarketPlace, and each column contains distinct information about the loan, such as loan terms, credit scores, borrowers' personal information etc.
See a brief summary of the columns in the dataset below:
Loan-related columns: LoanOriginalAmount, interest rate, Loan term, etc.
Borrower-related columns: income range, employment status, occupation, state etc.
Loan performance columns: current loan status, loan duration, delinquency status, etc.
Credit score columns: borrower's credit score, creditscore range etc.
Other columns: listing category, lender yield, etc.
Ultimately, the dataset is a structured dataset, with every of the 113,937 rows representing a single loan and each column representing a particular feature of that loan. 

### The main feature(s) of interest in this dataset

> Loan amount: the amount of money loaned to the borrower.
Interest rate: the annual interest rate charged on the granted loan .
Term length: the length of time/period the borrower is given to repay the loan.
Loan status: the current status of the loan (e.g., current, charged off, default, etc.).
Borrower's credit score: the credit score of the borrower at the time the loan was issued.
Income range: of the borrower.
Occupation: the occupation of the borrower.

### The features in the dataset that will help support my investigation into my feature(s) of interest

## Bivariate Exploration


## Question :How is loan performance affected by borrower income level?

```python
performance_by_income = df.groupby(['IncomeRange', 'LoanStatus']).size().unstack()
performance_by_income.plot(kind='bar', stacked=True)
plt.title('Loan Performance by Income Range')
plt.xlabel('Income Range')
plt.ylabel('Count')
plt.show()
```

## Question : How is the loan interest rate affected by borrower credit score?

```python
sns.scatterplot(data=df, x='CreditScoreRangeLower', y='BorrowerRate')
plt.title('Interest Rate by Credit Score')
plt.xlabel('Credit Score')
plt.ylabel('Interest Rate')
plt.show()
```

## Question: What is the relationship between loan performance and credit score?

```python
sns.scatterplot(data=df, x='CreditScoreRangeLower', y='LoanStatus', hue='LoanStatus')
plt.title('Loan Performance by Credit Score')
plt.xlabel('Credit Score')
plt.ylabel('Loan Status')
plt.show()
```

## Question: How does loan performance vary by borrower employment status?

```python
performance_by_employment = df.groupby(['EmploymentStatus', 'LoanStatus']).size().unstack()
performance_by_employment.plot(kind='bar', stacked=True)
plt.title('Loan Performance by Employment Status')
plt.xlabel('Employment Status')
plt.ylabel('Count')
plt.show()
```

## Question: How does loan interest rate vary by loan term?

```python
sns.lineplot(data=df, x='Term', y='BorrowerRate')
plt.title('Interest Rate by Loan Term')
plt.xlabel('Term (months)')
plt.ylabel('Interest Rate')
plt.show()
```

## Question: How does loan performance vary by the loan term

```python
sns.scatterplot(data=df, x='Term', y='LoanStatus', hue='LoanStatus')
plt.title('Loan Performance by Term')
plt.xlabel('Term')
plt.ylabel('Loan Status')
plt.show()
```

### Given the relationships observed in this part of the analysis, this is how the feature of interest varied with other features in the dataset:-

> In the bivariate part of the analysis, I analyzed a number of relationships between the feature of interest (loan performance) and other features in the dataset.
Firstly, I looked at Loan Performance by Income Range, and saw that the higher the income range, the higher the loan performance. That is higher income earners were almost twice likely to complete their loans.
I also observed that there was a strong positive correlation between loan performance and the credit score of the borrower. This isn't far-fetched because lenders usually see borrowers with higher credit scores as less risky and so they are more likely to be approved for loans. Ideally, borrowers with higher credit scores are expected to be more financially responsible.
Again,I observed that there were differences in Loan Performance by Employment Status . For example, loans taken out by full-time earners accounted for the best loan performance although these set still accounted for the most defaulters. 
Finally, I found that there was no major correlation between loan performance and the loan term. Both Loans with longer and shorter terms had similar loan performances.
Overall, these observations suggest that loan performance is influenced by a combination of factors, including the borrower's credit score, the loan term, and the purpose of the loan. By understanding these relationships, lenders may be better able to assess the risk of potential borrowers and make more informed lending decisions.






###  Relationships between the other features (not the main feature(s) of interest):-

>  I found that there was a negative correlation between the borrower's credit score and interest rate on the loan. This suggests that borrowers with higher credit scores may be able to negotiate lower interest rates with lenders, possibly because they are seen as less risky.
Another intriguing relationship I observed was that of the Interest Rate and the Loan Term, as the loan term increases, the interest rate slightly lowers.
These relationships suggest that the various features in the dataset depend on one another, and that multiple factors may influence borrowers' loan amounts, interest rates, and loan performance.  When lenders understand these relationship, they can then make more informed lending decisions and minimize risks.
